# Linear Regression betweeen Price and Sqft_Living

```python
price = data.iloc[:,:1]
```

```python
sqft = data.iloc[:,1:2]
```

```python
plt.scatter(sqft,price)
plt.title("Price VS Sqft")
plt.xlabel("Sqft")
plt.ylabel("Price")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_test,x_train,y_test,y_train = train_test_split(sqft,price,test_size=1/3,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
plt.scatter(x_train,y_train,color="blue")
plt.plot(x_train,model.predict(x_train),color = "black")
plt.title("Price Variations")
plt.xlabel("Sqft")
plt.ylabel("price")
```

```python
y_pred = model.predict(x_test)
```

```python
plt.scatter(x_test,y_test,color = "green")
plt.plot(x_test,model.predict(x_test),color = "red")
plt.title("Price Variations")
plt.xlabel("Sqft")
plt.ylabel("price")
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,model.predict(x_test)))
```

```python
r_square = r2_score(y_test,model.predict(x_test))
```

```python
print("THE RMSE Value is  :-",RMSE)
print("THE R_Square Value is   :-",r_square)
```
