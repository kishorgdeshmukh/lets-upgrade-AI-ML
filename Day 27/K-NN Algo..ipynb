```python
import pandas as pd
```

```python
from sklearn import preprocessing 
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
data = pd.read_csv("train.csv")
```

```python
data.isna().sum()
```

```python
data1 = data.fillna(method="bfill")
```

```python
le = preprocessing.LabelEncoder()
```

```python
data1["Sex"] = le.fit_transform(data["Sex"])
```

```python
data1["Embarked"] = le.fit_transform(data["Embarked"])
```

```python
data1
```

```python
y = data1["Pclass"]
```

```python
X = data1.drop(columns=["Name","Ticket","PassengerId","Cabin"],axis=1)
```

```python
X
```

```python
y
```

```python
from sklearn import neighbors
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
y_test.count()
```

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train).score(X_test,y_test)
```

```python
y_pred = knn.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
def KNN():
    y = data1["Pclass"]
    X = X = data1.drop(columns=["Name","Ticket","PassengerId","Cabin"],axis=1)
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
    z = y_test.count()
    
    for i in range(1,z):
        knn = neighbors.KNeighborsClassifier(n_neighbors=i)
        k.append(i)
        print("For K=",i,"Value")
        
        print("SCORE is ",knn.fit(X_train,y_train).score(X_test,y_test))
        
        y_pred = knn.predict(X_test)
        print("Confusion matrix")
        print(confusion_matrix(y_test,y_pred))
        print("=================\n")
        
    
```
